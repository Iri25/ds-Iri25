```python
# ex2
# a
reset()
t, k, T0, Tm = var('t, k, To, Tm')
T = function('T')(t)
eq = diff(T, t) == -k * (T - Tm)
f(t, k, T0, Tm) = desolve(eq, [T, t], [0, T0])
show(f(t, k, T0, Tm))
```

```python
# ex3
# a
reset()
t, r, x0, k, r0 = var('t, r, x0, k, r0')
x = function('x')(t)
deqM = diff(x, t) == r * x
sM(t, x0, r) = desolve(deqM, [x, t], [0, x0])
modMathus = show(sM(t, x0, r))
```

```python
# ex3
# a - cont

deqV1 = diff(x, t) == r0 * x * (1 - x/k)
sV1 = desolve(deqV1, [x, t], [0, x0])
modVerhulst = show(solve(sV1, x))

eq1 = sV1.log_simplify()
[sV2]= solve(eq1, x)
show([sV2])

sV(t, x0, r0, k) = sV2.rhs()
show(sV(t, x0, r0, k))
```
